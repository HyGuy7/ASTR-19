```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#create random data
n= 1000 #numer of random numbers to generate
#x = np.random.randn(n) #generate random data, from 0 to 1 with even distribution
x = np.random.random(n) #generate random data, from 0 to 1 with even distribution
#create histogram bins
width = 0.01
hist_min = np.floor(min(x)) #take the lowest values of x an dround it down
hist_max = np.ceil(max(x))+width #take the highest value of x and round it up

bins = np.arange(hist_min, hist_max, width) #this contains the borders of the bins

#quick look at what is happening
#print(bins)

#create the histogram
plt.hist(x, bins=bins, alpha=0.5,edgecolor='black')
plt.title('Session 9 Histogram')
plt.xlabel('x')
plt.ylabel('counts')
#save it
plt.savefig('session-9.pdf' , bbox_inches='tight', facecolor = 'white')

plt.show()
```
