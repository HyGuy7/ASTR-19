```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#generate a raleigh distribution of random data

a = 10
x = np.random.pareto(a, size=1000) #generate a pareto distribution with a shape of 10 and 1000 numbers
#create histogram bins
width = 0.01
hist_min = np.floor(min(x)) #take the lowest values of x an dround it down
hist_max = np.ceil(max(x))+width #take the highest value of x and round it up

bins = np.arange(hist_min, hist_max, width) #this contains the borders of the bins

#quick look at what is happening
#print(bins)

#create the histogram
plt.hist(x, bins=bins, alpha=0.5,edgecolor='black')
plt.title('Session 10 Histogram')
plt.xlabel('x')
plt.ylabel('counts')
#save it
plt.savefig('session-10.pdf' , bbox_inches='tight', facecolor = 'white')

plt.show()
```
