```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#make an array from 0 to 1 w/ 100 steps
x = np.linspace(0,1,100)
print(x)
#define function that returns exp(x)
def exponential(value):

    return np.exp(value)
```

```python
# set y=exp(x)
y = exponential(x)

#make a plot of x vs y
fig, ax = plt.subplots()
ax.plot(x, y)

#label the axis
plt.xlabel('Time [miliseconds]')
plt.ylabel('Awesomeness')

#save the figure
plt.savefig('awesomeness-plot.pdf' , bbox_inches='tight', facecolor = 'white')

plt.show()
```
