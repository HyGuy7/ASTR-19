```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#make an array from 0 to 1 w/ 100 steps
x = np.linspace(0,1,100)
print(x)

#define function that returns sin(x)
def sinx(value):

    return np.sin(value)
print
#define function that returns cos(x)
def cosx(value):

    return np.cos(value)
    
```

```python
#define the variables
y = sinx(x)
z = cosx(x)

#call subplots to make a multipanel figure
#this means 1 row, 2 columns figure
figure, axes = plt.subplots(1,2)

#first panel
axes[0].plot(x, y, color='blue')
axes[0].set_title('sin(x)')

#second panel
axes[1].plot(x, z, color='red')
axes[1].set_title('cos(x)')

#save the figure
plt.savefig('sine-and-cosine.pdf', bbox_inches='tight', facecolor = 'white')

plt.show()

```
